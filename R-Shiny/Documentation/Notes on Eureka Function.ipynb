## 1.Sample Size Estimation - Proportion Variable (Binary Outcomes)

In the case of sample size estimation, the formula used to calculate the sample size is:


$$ n = \frac {r + 1}{1} \frac {\bar p (1 - \bar p) (Z_\beta + Z_{\alpha/2})^2}{(p_1 - p_2)^2} $$

Where:

$n:$ size of the smaller group <br>
$r:$ ratio of larger group to smaller group <br>
$\bar p :$ Sum of $p_1$ sum lift <br>
$p_1 - p_2:$ Effect size to be measured <br>
$Z_\beta :$ Z critival value for the power level i.e. .84 = 80% <br>
$Z_{\alpha/2}:$ Z Critical value for two-tailed significance level i.e. 1.96 @ $\alpha = .05$


## 1.1. Simple Calculation of the Sample Size

Below, I provide a simple function that will calculate the the sample size given the components provided to the function corresponding to the formular above. Notice that we will need to use the qnorm function to return the critical values associated with Power and Significance level.

```python
import math
import numpy as np
from scipy.stats import norm

norm.ppf(.8), norm.ppf(.025)
```

```python
from scipy.stats import norm

def sample_calculation(ratio, p_1, lift, power, sig_level):
    """ Computing the Sample by Group """
    
    new_ratio = round(ratio*10/(1-ratio)/10, 2)
    
    z_power = norm.ppf(power)
    z_siglevel = abs(norm.ppf(sig_level/2))
    p_bar = p_1 + lift   
    
    part_1 = (new_ratio + 1)/new_ratio
    part_2 = ((p_bar)*(1 - p_bar)*(z_power + z_siglevel)**2)/(lift)**2
    sample = part_1*part_2
    
    return int(round(sample))
```

### 1.1.1 Compute Sample 

Now, level's compute the sample size estime for a 50-50 split of two campaigns with baseline of .5, lift size of 10% (.1) and power level 70% (.7) with a significance level of 5% (.05)

```python
sample_calculation(.5, .5, .05, .8, .05)
```

Given the above formular can take all the user inputs and return the sample estimates, it can be used interchangeably to estimate sample sizes based on various scenarios. 

A few things to point out:

1. We expect the ratio input to be entered in percentage and not in real values, therefore we can update the formulat to fix that. See below
2. The above sample is for the smaller group and is only equal to the larger group if the splits are meant to be equal

Fixing the ratio input

```python
def ratio_fix(ratio):
    
    return round(ratio*10/(1-ratio)/10, 2)
```

Tring the function above

```python
x = [.5, .6, .7, .75, .8, .9, .95]
split = [ratio_fix(x) for x in x]

x, split
```

We see that the function above works the way we want it to work. However, to be sensible about the splits, we need to limit our inputs between .5 and .9. That is we are willing only estimate based on the interval 50-50 through 90-10. The algorithm with allow for all intervals but its important to be conscious of the splits relative to the business problem.

## 1.2. Sample Size Estimation - Countinuous Variable Outcome

In the above example, we have seen the estimation of sample size from a binary outcome. We can also do the same for the continuous variable outcome. The general formular we will use here is the following:


$$ n = \frac {r + 1}{r} \frac {\sigma^2 (Z_{beta} + Z_{\alpha/2})^2 } { difference^2 } $$


Where:

$n:$ size of the smaller group <br>
$r:$ ratio of larger group to smaller group <br>
$\sigma :$ standard-deviation of the characteristic <br>
$difference:$ differences in means of the outcome <br>
$Z_\beta :$ Z critival value for the power level i.e. .84 = 80% <br>
$Z_{\alpha/2}:$ Z Critical value for two-tailed significance level i.e. 1.96 @ $\alpha = .05$

We can use the above formula for estimating univariate sample needed for detecting lift.

In the example below, we develop the code for providing this sample estimation.

```python
def ContinuousSampleEstimation(variable_series, mean_diffence, power, significance, ratio):
    
    """ This function will compute the sample Estimates for Continuours Variables"""
    z_power = norm.ppf(power)
    z_alpha = abs(norm.ppf(significance/2))
    variance = np.var(variable_series)
    
    ratio_part = (ratio + 1)/ratio
    sample_part = variance*(z_power + z_alpha)**2/(mean_diffence**2)
    
    return math.ceil(ratio_part*sample_part)
```

```python
ContinuousSampleEstimation([21.3, 31.1, 32.4, 18.2, 25.1, 23.9, 27.6, 34.01, 22.3], 3, .8, .01, 1)
```

## 1.3 Kolmogorov Smirnov Power Analysis - (Non-normal Continuous Data)

Most of our data may not be normally distribution or we may wish to test for continuous data points for non-normally distribution data. In this case, we would need to run a Kolmogorov Smirnov Simulation to determine what sample size required at every specified time frame. 

Below is the proposal and implementation of this.

1. Given some time series A (and say B), we can simulate what the time series would be in the next 4- six weeks
2. With every prediction/ we simulate and lift based on a random normal distribution with lift as the mean
3. Perform a Kolmogorov Smirnov Test after the simulations to determine the lift if impact
4. Look at the proportion of simulations the determined significance and convert to power proportion

Let's examine an example of the Generic Causal Impact Test Analysis using some data from Chilli's Test. Specification for Causal Impact is:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from causalimpact import CausalImpact

%matplotlib inline
```

```python
data = pd.read_csv('CausalImpactTest_Demo.csv', parse_dates=['Date'])
data.set_index('Date', inplace=True)
data.head()
```

Now let's define a function that will filter the data by metric and return a pivot 

```python
def filterPivot(datafile, metric):
    
    filtered_df = datafile.loc[:, ['Group' , metric]].copy()
    filtered_df = filtered_df.pivot_table(index=filtered_df.index, values=metric ,columns=['Group'] )
    
    return filtered_df
```

```python
sessions = filterPivot(data, 'Sessions')
sessions.head()
```

```python
sessions.plot(figsize=(12,7))
```

#### Building the Causal Impact Model

Now that we have our metrics, we can build the causal impact model. In order to build our Causal Impact model, we need to specify our pre and post periods.
We will use the index names for this


```python
pre_period = [min(sessions.index), '2017-06-20']
post_period = [ '2017-06-21' ,max(sessions.index) ]
```

```python
sessions.head()
```

```python
impact = CausalImpact(sessions[['Test', 'Control']], pre_period=pre_period, post_period=post_period, alpha=.1)
impact.run()
```

```python
impact.inferences.head()
```

```python
impact.plot()
```

```python
impact.summary()
```
